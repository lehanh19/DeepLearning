# Dự đoán xem một khoản vay có được đồng ý cho vay hay không

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# Load data
data = pd.read_csv('dataset.csv').values
N, d = data.shape
x = data[:, 0:d-1].reshape(-1, d-1)
y = data[:, d-1].reshape(-1, 1)
```

```python
# Plot data
plt.scatter(x[:10, 0], x[:10, 1], c = 'red', edgecolors='none', s=30 , label='cho vay')
plt.scatter(x[10:, 0], x[10:, 1], c = 'blue', edgecolors='none', s=30 , label='từ chối')
plt.legend(loc=1)
plt.xlabel('Mức lương (triệu)')
plt.ylabel('Kinh nghiệm (năm)')
```

```python
# add column 1 into x data
x = np.hstack((np.ones((N, 1)), x))

w = np.array([0., 0.1, 0.1]).reshape(-1, 1)
```

```python
w = np.array([0., 0.1, 0.1]).reshape(-1, 1)
```

```python
numOfIteration = 1000
cost = np.zeros((numOfIteration, 1))
learning_rate = 0.01
```

```python
for i in range(1, numOfIteration):
    y_predic = sigmoid(np.dot(x, w))
    cost[i] = -np.sum(np.multiply(y, np.log(y_predic)) + np.multiply((1 - y), np.log(1 - y_predic)))
    w = w - learning_rate*np.dot(x.T, y_predic - y)
    print(cost[i])
```

```python
y_predict = sigmoid(np.dot(x, w))
```

```python
print('y_predic', y_predict[19])
print('y', y[19])
```

Gọi t (0<t<1) là xác suất cho vay tiền khi đó:

y_predict > t

<=> sigmoid(w0 + w1*x1(i) + w2*x2(i)) > t

<=> w0 + w1*x1(i) + w2*x2(i) > -ln(1/t - 1)

=> Đường phân cách x2(i) = - (w0 + w1*x1(i) + ln(1/t - 1)) / w2

```python
t = 0.5
plt.scatter(x[:10, 1], x[:10, 2], c = 'red', edgecolors='none', s=30 , label='cho vay')
plt.scatter(x[10:, 1], x[10:, 2], c = 'blue', edgecolors='none', s=30 , label='từ chối')
plt.plot((4, 10), (-(w[0]+4*w[1] + np.log(1/t - 1))/w[2], -(w[0] + 10*w[1] + np.log(1/t - 1))/w[2]), 'g')
plt.legend(loc=1)
plt.xlabel('Mức lương (triệu)')
plt.ylabel('Kinh nghiệm (năm)')
plt.show()
```
